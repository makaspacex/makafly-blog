```python
from pathlib import Path
import re
import html
```

```python
# bash 教程的生成

with open("runtime/bash-tutorial/chapters.yml") as f:
    content= f.read()

result = list(re.finditer(r"^- (.*?): (.*)", content, flags=re.MULTILINE))

frontmater_tp="""---
title: {}
lang: zh-CN
hidden: true
date: 2024-06-03 14:30:00
author: 阮一峰
recommend: false
cover: 
tags:
- bash
---

"""
menu_all_tp = \
"""
export default function sidebarBash() {{
  return [
      {{
      items: [\n {} \n]
      }}
   ]
}}
"""

items = ""
for i, ele in enumerate(result):
    
    fname,title = ele.group(1), ele.group(2)
    
    f_path = Path(f"runtime/bash-tutorial/docs/{fname}")
    front = frontmater_tp.format(f"{title}",i+1)
    pp = Path("docs")
    sub_path = Path(f"programer/bash/{(i+1):02d}-{f_path.name}")
    new_f_path = pp / sub_path
    new_f_path.parent.mkdir(exist_ok=True)
    
    with open(f_path, 'r') as f:
        content = f.read()
        content =  front + content
        
    with open(new_f_path, 'w+') as f:
        f.write(content)
    items += f"        {{ text:'{i+1}.{title}', link: '{sub_path.parent/sub_path.stem}' }},\n"

outline = menu_all_tp.format(f"{items}")
with open("docs/.vitepress/sidebar/bash.ts", 'w+') as f:
    f.write(outline)
```

```python

# javascript教程的生成
with open("runtime/javascript-tutorial/chapters.yml") as f:
    content= f.read()

result = list(re.finditer(r"^- (.*?.md): (.*)", content, flags=re.MULTILINE))
h1_list = list(re.finditer(r"^- (.*?/): (.*)", content, flags=re.MULTILINE))

outline_map = {}


frontmater_tp="""---
title: {}
lang: zh-CN
hidden: true
date: 2024-06-03 15:30:00
author: 阮一峰
recommend: false
cover: 
tags:
- javascript
---

"""

menu_all_tp = \
"""
export default function sidebarJavascript() {{
  return [
      {}
   ]
}}
"""

menu_item_tp = \
"""    {{
      text: '{}', 
      collapsed: false,
      items: {},
    }},
"""

menu_item = ""

for i_h1, h1_ele in enumerate( h1_list):
    h1_path, h1_name = h1_ele.group(1),h1_ele.group(2)
    sub_list = []
    for ele in result:
        fname = ele.group(1)
        if not fname.startswith(h1_path):
            continue 
        sub_list.append(ele)
    items = ""
    for i, ele in enumerate(sub_list):

        fname,title = ele.group(1),ele.group(2)
        title = html.escape(title)
        
        
        # print(fname, title)
        f_path = Path(f"runtime/javascript-tutorial/docs/{fname}")
        front = frontmater_tp.format(f"{title}",i+1)
        
        pp = Path("docs")
        sub_path = Path(f"programer/javascript/{(i_h1+1):d}.{(i+1):d}-{f_path.name}")
        new_f_path = pp / sub_path
        # new_f_path = Path(f"runtime/dd/{(i_h1+1):d}.{(i+1):d}-{f_path.name}")
        new_f_path.parent.mkdir(exist_ok=True)
        
        with open(f_path, 'r') as f:
            content = f.read()
            content =  front + content
        
        with open(new_f_path, 'w+') as f:
            f.write(content)

        
        
        # { text: '介绍', link: '/algorithm/hash-table/' },
        items += f"        {{ text:'{i_h1+1}.{i+1} {title}', link: '{sub_path.parent/sub_path.stem}' }},\n"
    items = f"[\n{items}      ]"
    menu_item += menu_item_tp.format(f"{i_h1+1}.{h1_name}", items)

outline = menu_all_tp.format(menu_item)
with open("docs/.vitepress/sidebar/javascript.ts", 'w+') as f:
    f.write(outline)
```
